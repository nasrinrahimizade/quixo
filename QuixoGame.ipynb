```python
def simulate_games(num_simulation=100):
        my_player_wins = 0
        for _ in range(num_simulation):
            game = Game()  # Initialize a new game for each simulation
            player1 = MyPlayer(num_simulations=10)  # MCTS player #my player id is 0
            player2 = RandomPlayer()  # Random player as the opponent
            winner = game.play(player1, player2)  # Play out the game
            if winner == 0:  # Check if MyPlayer (assuming player1 is MyPlayer and wins as player 0)
                my_player_wins += 1
        return my_player_wins

# Run the simulation
num_simulation = 100
my_player_wins = simulate_games(num_simulation)
winrate_monte_carlo = my_player_wins / num_simulation
print(f"MyPlayer won {my_player_wins} out of {num_simulation} games.")
```

```python
print('win rate minimax' , winrate_minimax)
print('winrate_alpha_beta', winrate_alpha_beta)
print('win rate monte carlo', winrate_monte_carlo)
```

```python
import matplotlib.pyplot as plt

# Data
percentages = [winrate_minimax, winrate_alpha_beta, winrate_monte_carlo]
labels = ['minimax', 'alpha_beta', 'monte carlo']
colors = ['red', 'green', 'blue']

# Plotting the bar chart
plt.bar(labels, percentages, color= colors)

# Adding titles and labels
plt.title('Percentage Distribution')
plt.xlabel('models')
plt.ylabel('win rate')

# Display the plot
plt.show()
```
